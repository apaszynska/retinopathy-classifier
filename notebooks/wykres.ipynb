```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 1. Twoje dane
data = {
    'Class': ['Healthy (Grade 0)', 'Mild (Grade 1)', 'Moderate (Grade 2)', 'Severe (Grade 3)'],
    'Count': [547, 149, 237, 249]
}
df = pd.DataFrame(data)

# 2. Ustawienia wykresu (usunięta siatka i zmieniony rozmiar)
sns.set_style("white") # Zmiana stylu na "biały" usuwa linie siatki
plt.figure(figsize=(10, 6))

# 3. Stworzenie wykresu słupkowego z nową paletą kolorów
# Paleta kolorów z odcieniami niebieskiego i różu
custom_palette = ['#4682B4', '#FF69B4', '#1E90FF', '#C71585'] # SteelBlue, HotPink, DodgerBlue, MediumVioletRed
ax = sns.barplot(x='Class', y='Count', data=df, palette=custom_palette)

# 4. Dodanie tytułu i etykiet osi
ax.set_title('Class Distribution in the Original Dataset', fontsize=16)
ax.set_xlabel('Disease Stage', fontsize=12)
ax.set_ylabel('Number of Images', fontsize=12)
plt.xticks(rotation=15)

# 5. Dodanie dokładnej liczby nad każdym słupkiem
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', 
                (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha = 'center', va = 'center', 
                xytext = (0, 9), 
                textcoords = 'offset points')

# 6. Wyczyszczenie i wyświetlenie gotowego wykresu
plt.tight_layout()
sns.despine() # Opcjonalnie: usuwa górną i prawą ramkę dla czystszego wyglądu
plt.show()
```
